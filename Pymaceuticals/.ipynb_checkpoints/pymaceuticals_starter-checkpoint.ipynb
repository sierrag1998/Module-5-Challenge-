## Observations and Insights 

The data displays 249 different mice and their tumor growth/shrinkage with differnt drug regimens. 

Through analysis, several observations were made. 

First, Capomulin had the most timepoints from all the mice tested. 

Secondly, mice on the Ketapril drug regimen had the highest mean and median tumor volumes compared to all other mice on different drug regimens. 

Third, of all the mice tested, 50.7% were male mice and 49.35% were female mice. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

mice_df = pd.merge(mouse_metadata, study_results,
                                 how='outer', on='Mouse ID')

# Display the data table for preview
mice_df.head(10)

```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.


#GroupBys
mean_mice = clean_mice_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
median_mice = clean_mice_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
std_mice = clean_mice_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
var_mice = clean_mice_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
sem_mice = clean_mice_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

#Merging 
merge=pd.merge(mean_mice,median_mice, on="Drug Regimen")
merge_2=pd.merge(merge,std_mice, on="Drug Regimen")
merge_3=pd.merge(merge_2,var_mice, on="Drug Regimen")
merge_4=pd.merge(merge_3,sem_mice, on="Drug Regimen")

#Clean and Display 
merge_4.columns = ["Mean of Tumor Volume (mm3)", "Median of Tumor Volume (mm3)", "Standard Deviation of Tumor Volume (mm3)", "Variance of Tumor Volume (mm3)","SEM of Tumor Volume (mm3)"]
merge_4

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

#Get info 
info_df = clean_mice_df[["Drug Regimen","Tumor Volume (mm3)"]]
info_df_rename = info_df.set_index("Drug Regimen")
info_df_rename = info_df_rename.groupby(["Drug Regimen"])["Tumor Volume (mm3)"]

#Agg
summary_stat = info_df_rename.agg(['mean', 'median',"std","var","sem"])

#Clean and Display 
summary_stat = summary_stat.rename(columns={"mean": "Mean of Tumor Volume (mm3)", "median": "Median of Tumor Volume (mm3)", "std" : "Standard Deviation of Tumor Volume (mm3)","var":"Variance of Tumor Volume (mm3)","sem":"SEM of Tumor Volume (mm3)"})
summary_stat
```

```python
#Reduce DF 
treatment_df = q_merge.loc[(q_merge["Drug Regimen"] == "Capomulin") | (q_merge["Drug Regimen"] == "Ramicane") | (q_merge["Drug Regimen"] == "Ceftamin") | (q_merge["Drug Regimen"] == "Infubinol"),  :]
treatment_df

#Find data with last timepoint
match_loc = treatment_df.loc [treatment_df["Timepoint"] == treatment_df["Last Timepoint for Mouse ID"]]

    
matched_tumor =match_loc["Tumor Volume (mm3)"]  


 
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Determine outliers using upper and lower bounds


#Bounds Calculations 
quartiles = matched_tumor.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)


#Outlier Calculations 
outliers = match_loc.loc[(match_loc["Tumor Volume (mm3)"] > 94.04) | (match_loc["Tumor Volume (mm3)"] < 3.07),:]


#Print Statements 
print(f"The quartiles of the tumor(mm3) are: {quartiles}")
print(f"The interquartile range of the tumor(mm3) is: {iqr}")
print(f"The lower bound of the tumor(mm3) is: {lower_bound}")
print(f"The upper bound of the tumor(mm3) is: {upper_bound}")
print(f"There were no outliers found.")

```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest


campomulin_data = match_loc.loc[(match_loc["Drug Regimen"] == "Capomulin"),:]
campomulin_tumors = campomulin_data["Tumor Volume (mm3)"]
ramicane_data = match_loc.loc[(match_loc["Drug Regimen"] == "Ramicane"),:]
ramicane_tumors = ramicane_data["Tumor Volume (mm3)"]
infubinol_data = match_loc.loc[(match_loc["Drug Regimen"] == "Infubinol"),:]
infubinol_tumors = infubinol_data["Tumor Volume (mm3)"]
ceftamin_data = match_loc.loc[(match_loc["Drug Regimen"] == "Ceftamin"),:]
ceftamin_tumors = ceftamin_data["Tumor Volume (mm3)"]

match_loc["campomulin_tumors"] = campomulin_tumors
match_loc["ramicane_tumors"] = ramicane_tumors
match_loc["infubinol_tumors"] = infubinol_tumors
match_loc["ceftamin_tumors"] = ceftamin_tumors


# drug_merge=pd.merge(campomulin_data,ramicane_data, on="Mouse ID")

fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume of each Mouse across Four Regimens of Interest')
ax1.set_ylabel('Tumor Volume (mm3)')
match_loc.boxplot(column=["campomulin_tumors", "ramicane_tumors","infubinol_tumors","ceftamin_tumors"],rot=90)  
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient for mouse weight and tumor volume for the Capomulin regimen

#Calculations 
tumor = capomulin_data["Tumor Volume (mm3)"]
weight = capomulin_data["Weight (g)"]
cc = {round(st.pearsonr(tumor,weight)[0],2)}

#Print Line 
print(f"The Correlation Coefficient for mouse weight and tumor volume for the Capomulin regimen is {cc}")
```

```python
# Calculate the linear regression model for mouse weight and tumor volume for the Capomulin regimen


mouse_slope, mouse_int, mouse_r, mouse_p, mouse_std_err = linregress(tumor, weight)

mouse_fit = mouse_slope * tumor + mouse_int

#Build Model 
plt.scatter(tumor,weight)
plt.plot(tumor,mouse_fit,"--")
plt.title("Linear Regression Model for Mouse Weight and Tumor Size (mm3) on Capomulin ")
plt.xlabel("Tumor Volume (mm3)")
plt.ylabel("Weight (g)")
plt.figure(figsize=(10,8))
plt.show()
```
